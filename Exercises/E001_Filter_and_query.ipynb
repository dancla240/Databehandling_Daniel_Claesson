**Q3:** What is the length, width and height of the most expensive car in the entire dataset?

<details>
<summary>Answer</summary>
<br>&nbsp;&nbsp;&nbsp;length=199.2
&nbsp;&nbsp;&nbsp;width=72.0
&nbsp;&nbsp;&nbsp;height=55.4

```python
# För den dyraste bilen, ta fram width, height, length:
# använder loc + index för den dyraste bilen, som jag får fram med idxmax() metoden.
width = autos.loc[autos['price'].idxmax(),'width']
height = autos.loc[autos['price'].idxmax(), 'height']
length = autos.loc[autos['price'].idxmax(), 'length']
print(f'Length is {length}, height is {height}, width is {width}, of the most expensive car.')
```
