### gör en csv fil av autos
- index=None gör att indexen inte skrivs till csv filen, vilket man ofta inte vill ha med
- alla dataframes har index

```python
autos.to_csv("../Data/autos.csv", index=None)
```

### skriva till .json
- default skriver den kolumn för kolumn
- om man skriver orient="records" skriver den rad för rad.

```python
autos.to_json("../Data/new_autos.json", orient="records")
```

```python
json_string = autos.to_json(indent=4, orient="records")
print(json_string[0:200])
```

### till dictionary
- ger en lista av dictionaries

```python
my_dict = autos.to_dict(orient="records")
print(my_dict[0])
```

### XML

- liknande .json, fyller liknande syfte
- XML används tex för vektorgrafik, som .svg filer, man kan ladda ner sådan fil och kolla hur den ser ut
- [Svenska flaggan som .svg](https://upload.wikimedia.org/wikipedia/commons/4/4c/Flag_of_Sweden.svg)
