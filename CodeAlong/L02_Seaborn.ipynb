# Seaborn
- [Link to Seaborn doc](https://seaborn.pydata.org/index.html)
- templates för olika sorters grafer
- kolla API reference
- kolla in dataseten som man kan ladda ner  
[Länk till datasets](https://github.com/mwaskom/seaborn-data)
- figure: hela grafen
- axes: en av bilderna i figuren.
- Plots:
- relplot (relational):
- displot (distributions)
    1. histplot (histogram)
    2. KDE-plot (kernel density estimate)
    3. ECDF-plot (empirical cumulative distribution functions)
- catplot (categorical):
    - Categorical scatterplots:
        1. stripplot (liknande relational scatter plot, men sprider ut)
        2. swarmplot (liknande relational scatter plot, men sprider ut)
    - Categorical distribution plots:
        3. boxplot
        4. violinplot
        5. boxenplot
    - Categorical estimate plots:
        6. pointplot
        7. barplot
        8. countplot
- Om man vill plotta flera plottar i samma plot (subplots) ska man använda 'relplots' tex. Om man bara vill plotta en scatter, kan man använda 'scatter' direkt.
![Types of plots](..\Assets\function_overview_8_0.png)

```python
import seaborn as sns
import warnings

warnings.filterwarnings("ignore", "is_categorical_dtype")
warnings.filterwarnings("ignore", "use_inf_as_na")
warnings.filterwarnings("ignore", "The figure layout has changed to tight")
warnings.filterwarnings("ignore", "The default of observed=False is deprecated and will be changed to True in a future version of pandas. Pass observed=False to retain current behavior or observed=True to adopt the future default and silence this warning.")
```

```python
tips = sns.load_dataset('tips')
tips.head() #tips.tail(), 5 default
```

## Relational (relplot)

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', hue='sex', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', hue='sex', style = 'smoker', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', hue='sex', size ='size', style = 'smoker', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', hue='sex', size ='size', sizes=(10,60), style = 'smoker', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', hue='sex', size ='size', sizes=(10,60), row='day', style = 'smoker', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', hue='sex', size ='size', sizes=(10,60), col='day', height=3, style = 'smoker', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', hue='sex', size ='size', sizes=(10,60), col='day', row='time', height=3, style = 'smoker', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', hue='sex', size ='size', sizes=(10,60), col='day',col_wrap=2, height=3, style = 'smoker', data=tips)
```

## Categorical (catplot)
- Byter ut 'total bill" till 'day' som är categorical.
- Denna första plot är inte så användbar.

```python
sns.relplot(x = 'day', y = 'total_bill', data = tips)
```

Byter till 'catplot' från 'relplot'
- stripplot (sprider ut värdena i x-led).

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips).set(title = 'stripplot')
```

Swarmplot

```python
sns.catplot(x = 'day', y = 'total_bill', kind='swarm', data = tips).set(title = 'swarmplot')
```

Boxplot

```python
sns.catplot(x = 'day', y = 'total_bill', kind='box', data = tips).set(title = 'boxplot')
```

```python
sns.catplot(x = 'day', y = 'total_bill', kind='violin', data = tips).set(title = 'violinplot')
```

```python
sns.catplot(x = 'day', y = 'total_bill', kind='boxen', data = tips).set(title = 'boxenplot')
```

Boxplot som är vriden

```python
sns.catplot(y = 'day', x = 'total_bill', hue = 'sex', kind='box', data = tips).set(title = 'boxplot')
```

```python
sns.catplot(x = 'day', y = 'total_bill', hue = 'sex', kind='bar', data = tips).set(title = 'barplot')
```

'Count' räknar per categori

```python
sns.catplot(x = 'day', hue = 'sex', kind='count', data = tips).set(title = 'count plot')
```

## Distribution (displot)
- kind = 'hist' är default
- man väljer antal 'bins'

```python
sns.displot(x='total_bill', data=tips)
sns.displot(x='total_bill', bins=50, data=tips)
```
